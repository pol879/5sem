# Task1
Вам на вход приходит большая матрица из целых чисел. Логически это карта. Каждый элемент матрицы - клеточка карты. Значение элемента - высота над уровнем моря в данной клеточке (если число положительное) или глубина моря (если число отрицательное). Размеры карты - до 1024 по каждой из сторон.

Необходимо найти:

общую площадь клеточек моря, в которых его глубина больше 5; 

общий объём всей воды на карте; 

максимальную высоту над уровнем моря, которая есть на этой карте.


Формат входных данных 

В первой строке 2 целых числа - N и M. Далее N строк, в каждой по M целых чисел через пробел.

(Внимание: кто-то особо добрый отформатировал ввод так, что его удобно читать глазами. Но в итоге между соседними числами бывает разное количество пробелов. С этим придётся справиться.)

Формат выходных данных 

Три строки с целыми числами. В каждой - ответ на соответственный вопрос.

Примеры

Ввод 

4 6

 1  4  4  2  1  1 
 
 0  0  1  0 -1 -1 
 
-2 -3 -1 -5 -6 -4 

-1 -3 -3 -4 -4 -2 

Вывод  

1 

40

4 
      


```python
n, m = map(int, input().split())
a=np.zeros(n*m)
c=0
for i in range(n):
    a[c:c+m]=list(map(int, input().split()))
    c+=m
b= pd.DataFrame(a)
ans1 = (b[0]<-5).astype(int).sum()
ans2 = int(abs((b[0]<0).astype(int) @ b[0]))
ans3 = int(list(b[b[0]>0].max())[0])
print(ans1, ans2, ans3, sep='\n')
```

# Task2

Реализуйте одну очень странную нейросеть. Нейросеть должна:

принимать на вход вектор признаков (чисел с плавающей точкой); \
в качестве первого преобразования умножать вектор признаков на матрицу весов A2 (на выходе получается новый вектор); \
в качестве второго преобразования умножать вектор, полученный на прошлом шаге, на вектор весов b (на выходе получается скалярное число). \
Размерности матриц и векторов - в пределах 1024. \
**Формат входных данных** \
В первой строке приходит путь к файлу, в котором хранится матрица весов A размерности NxM. Файл содержит N строк в каждой по M значений через пробел. Каждое значение - число с плавающей точкой.

Чтобы получить матрицу A2 для первого преобразования нужно возвести данную матрицу A в квадрат.

Во второй строке приходит путь к файлу, в котором хранится вектор весов b. Файл содержит одну строку с N значениями через пробел. Каждое значение - число с плавающей точкой.

В третьей строке приходит входной вектор X. Формат строки - M значений с плавающей точкой через пробел.

**Формат выходных данных** \
Одно число - результат прохождения вектора X через нашу странную нейросеть. Число выводить со всеми знаками после запятой, какие получатся.

```python
path_1 = input()
path_2 = input()
X = np.array(list(map(float, input().split())))

A = np.array(pd.read_csv(path_1, sep="\s+", header=None))
A2 = np.matmul(A, A)
b = np.array(pd.read_csv(path_2, sep="\s+", header=None))

res1 = X.dot(A2)
res2 = b@(A2@X)

print(res2[0])
```

# Task4

Одна из метрик, которой принято меряться в науке - цитируемость. Вам даны данные набора научных статей. По каждой статье известны название, автор, какие другие статьи цитирует данная статья. \

Для каждой статьи подсчитайте, сколько раз её процитировали другие статьи из этого набора. \
Для каждого автора подсчитайте среднее количество цитирований по всем статьям данного автора в данном наборе. (То есть, если у автора 2 статьи, из которых одну процитировали 5 раз, а вторую 0 раз, то его средняя цитируемость 2.5.) \

Выведите 3 наиболее цитируемых авторов в порядке убывания среднего количества цитирований. \
**Формат входных данных** \
В первой строке приходит путь к csv-файлу, в котором хранятся данные о статьях. По каждой статье в файле содержатся её идентификационный номер, название, автор. (В реальной жизни всё заметно сложнее. Как минимум, авторов у статьи несколько. Но в этой задаче всё вот так упрощённо.)

Во второй строке приходит путь к файлу, в котором хранятся данные о том, кто кого цитирует. В файле дано множество ссылок между статьями в формате кто ссылается (поле paper_id, в нём идентификационный номер статьи) и на кого ссылается (поле reference, в нём название статьи, на которую идёт ссылка). Иногда ссылки идут на статьи, которых нет в данном наборе - это нормально. (В реальной жизни эти данные очень грязные. Но в данной задаче всё аккуратно, опечаток в ссылках нет.)

**Формат выходных данных** \
В трёх строках 3 самые цитируемые автора. У каждого вывести имя, после него через пробел среднее количество цитирований статей (строго с тремя знаками после запятой).

```python
papers=pd.read_csv('papers001.csv', sep=';')
links = pd.read_csv('links001.csv', sep=';')
```

```python
papers=pd.read_csv(input(), sep=';')
links = pd.read_csv(input(), sep=';')
papers['count'] = np.zeros_like(papers['id'])
papers.index = list(papers['title'])
papers_set = set(papers['title'])
for paper in links['reference']:
    if paper in papers_set:
        papers.loc[paper,'count']+=1
all_p = papers.value_counts(subset = 'author')

papers['hirsh']=np.zeros_like(papers['count'], dtype = float)
for row in papers.index:
    papers.loc[row, 'hirsh'] = papers.loc[row, 'count']/all_p[papers.loc[row, 'author']]

ans = (
    papers
    .groupby('author')
    .agg({'hirsh': 'sum'})
).sort_values('hirsh', ascending=False).head(3)
for row in ans.index:
    print(row, '{:.3f}'.format(round(ans.loc[row, 'hirsh'], 3)))
```

# Task 5

Задано прямоугольное поле из квадратных клеточек. Размер поля N x M клеточек. В начальный момент времени на поле есть пустые клеточки (отмечены символом '.') и обитаемые клеточки (отмечены символом '#').

Выполняется K шагов игры. На каждом шаге каждая клеточка взаимодействует со своими соседями. Соседями считаются клеточки, прилегающие по горизонтали, вертикали и диагонали. (Т.е. максимум 8 соседей, на краях поля меньше.)

Правила взаимодействия на каждом шаге такие:

Если у живой клетки меньше двух живых соседей, то она погибает.
Если у живой клетки два или три живых соседа, то она выживает.
Если у живой клетки больше трёх живых соседей, то она погибает.
Если у неживой клетки ровно три живых соседа, она становится живой.
Альтернативная формулировка ровно тех же правил:

Если у живой клетки ровно два или три живых соседа, она остаётся живой.
Если у неживой клетки ровно три живых соседа, она становится живой.
Все остальные клетки умирают или остаются мёртвыми.
Сделайте K шагов игры, выведите итоговое поле после этих шагов.

**Формат входных данных** \
Три положительных целых числа N (размер поля по вертикали), M (размер поля по горизонтали), K (количество шагов).

Далее посимвольно вводится стартовое поле N x M, мёртвые клеточки на нём отмечены '.', живые - '#'.

**Формат выходных данных** \
Итоговое поле N x M после K шагов, клеточки отмечены ровно аналогично.

```python
N, M, K = map(int, input().split())
board = [list(input().strip()) for _ in range(N)]


def next_state(field, row, col):
    alive = np.sum(field[max(0, row - 1):min(N, row + 2), max(0, col - 1):min(M, col + 2)]) - field[row, col]   
    if field[row, col] == 1 and alive in [2, 3]:
        return 1
    elif field[row, col] == 0 and alive == 3:
        return 1
    else:
        return 0
        

def count(N, M, K, field0):
    '''Функция для расчета следующего состояния клеточек'''  
    field = np.array([[1 if x == '#' else 0 for x in row] for row in field0])
    for _ in range(K):
        next_field = np.zeros((N, M))
        for i in range(N):
            for j in range(M):
                next_field[i, j] = next_state(field, i, j)
        
        field = next_field
    result_field = [['#' if x == 1 else '.' for x in row] for row in field]
    return result_field


res = count(N, M, K, board)
for row in res:
    print(''.join(row))
```
