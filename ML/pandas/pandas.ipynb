# Task2:
Analyse dataset from kaggle

Link to dataset: https://www.kaggle.com/datasets/sukhmandeepsinghbrar/heart-attack-dataset 

```python
data = pd.read_csv('Medicaldataset.csv')
data
```

```python
data.isna().astype(int).sum()
```

```python
data.describe()
```

```python
data.columns = ['age', 'gender', 'heart_rate', 'sys_blood_pres', 'dias_blood_pres', 'blood_sugar', 'CK-MB', 'troponin', 'result']
data
```

```python
data.query('heart_rate > 200')
```

```python
high_sugar = data.query('blood_sugar>300')
low_sugar = data.query('blood_sugar< 80')
(
    low_sugar
    .groupby('result')
    .agg({'blood_sugar':'mean'})
)
```

```python
(
    high_sugar
    .groupby('result')
    .agg({'blood_sugar':'mean'})
)
```

```python
data.loc[:, 'result'].value_counts()
```

```python
sns.pairplot(data=data, hue='result')
```
